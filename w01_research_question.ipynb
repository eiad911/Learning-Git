## 1. My lane (or freestyle) and why

```python
## My Lane

I chose the **Content Refresh Prioritization** lane.

I selected this lane because websites often contain thousands of pages, making it difficult to decide which pages should be updated first. By using data and machine learning, we can prioritize pages that are most likely to be declining and need a content refresh. This can help teams spend their time on the pages with the greatest potential impact.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## Research Question

**Question:**
Can we identify web pages that should be prioritized for content refresh based on historical search performance and page characteristics?

**Decision:**
Which pages should be reviewed and updated first?

**Action:**
The SEO or content team will review the highest-priority pages and update their content.

**Cost of a Wrong Recommendation:**
If a declining page is missed, valuable search traffic may continue to decrease. If a healthy page is incorrectly prioritized, time and resources may be wasted.

**Why Machine Learning?**
Machine learning can combine multiple signals, such as impressions, CTR, page age, and average position, to rank pages more effectively than relying on simple manual rules.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd

if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("../..")

print("Current directory:", os.getcwd())

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))
print("Declining pages (%):", round((df["trend_direction"]=="down").mean()*100,2))

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation:", round(corr,3))
```

### Observations

- The dataset contains **30,000 pages**.
- About **54.2%** of the pages are labeled as declining.
- The correlation between search volume and impressions is approximately **0.001**, indicating a very weak relationship.

These observations suggest that multiple features should be considered together instead of relying on a single metric.
